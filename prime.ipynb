```python
import PyPDF2

# Otwrz plik PDF
with open("prime_intellect.pdf", "rb") as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    extracted_text = ""

    # Iteracja przez strony PDF
    for page in pdf_reader.pages:
        extracted_text += page.extract_text()  # Ekstrakcja tekstu z kadej strony

print(extracted_text)  # Wywietl wyekstrahowany tekst
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Zaaduj tokenizer i model
model_name = "gsarti/opus-mt-tc-en-pl"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# Fragment tekstu do przetumaczenia
text = "This is an example sentence to be translated into Polish."

# Tokenizacja tekstu
inputs = tokenizer(text, return_tensors="pt")

# Tumaczenie
translated = model.generate(**inputs)

# Dekodowanie przetumaczonego tekstu
translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

print(translated_text)
```

```python
def first_n_words(text, n):
    # Podziel tekst na sowa
    words = text.split()
    # Zwr pierwsze n sw
    return ' '.join(words[:n])

# Przykad uycia

n = 159

result = first_n_words(extracted_text, n)
print(result)  # Wyjcie: "To jest przykadowy tekst,"
```

```python
# Tokenizacja tekstu
inputs = tokenizer(result, return_tensors="pt")

# Tumaczenie
translated = model.generate(**inputs)

# Dekodowanie przetumaczonego tekstu
translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

print(translated_text)
```
